```python
import json
import requests
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36'}
```

```python
districts_df = pd.read_csv("districts.csv")
```

```python
#districts_list = ['Rangareddy','Hyderabad','Medchal','Sangareddy','Nalgonda']
districts_list = ['Rangareddy','Medchal','Sangareddy','Nalgonda','Hyderabad']
dates_list = ['27-05-2021'] #,'29-05-2021','30-05-2021','31-05-2021','01-06-2021'
```

```python
slots_df = pd.DataFrame()
for d in districts_df[districts_df['district_name'].isin(districts_list)]['district_id']:
    for da in dates_list:
        r = requests.get('https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/findByDistrict?district_id='+str(d)+'&date='+da,headers=headers)

        slots = pd.DataFrame(r.json()['sessions'])

        if not slots.empty:
            #slots = slots[slots['available_capacity_dose1']>0]
            slots_df = slots_df.append(slots)

    #print(slots[['name', 'vaccine','pincode','min_age_limit','available_capacity','available_capacity_dose1']])

columns_to_show = ['name', 'vaccine','pincode','district_name','min_age_limit','date','available_capacity','available_capacity_dose1']
slots_df = slots_df.loc[ slots_df['min_age_limit'] == 18,columns_to_show].sort_values(by=['available_capacity_dose1', 'district_name','name'])
slots_df
```
